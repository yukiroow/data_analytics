```python
import pandas as pd

shows = pd.read_csv("data/shows.csv")
shows
```

### Clean the dataset.
- Drop irrelevant columns
- Re-arrange columns
- Convert monetary values to integer values

```python
# Drop irrelevant columns
shows = shows.drop(columns=['Peak', 'All Time Peak', 'Ref.'])

# Re-arrange columns to better display data
shows = shows[['Rank', 'Tour title', 'Artist', 'Year(s)', 'Shows', 'Actual gross', 'Actual gross (in 2022 dollars)', 'Average gross']]

# Function to remove the dollar signs ($) and commas (,)
# Also type casts the cleaned value to an integer
def clean_monetary_value(num):
    if type(num) is None:
        return None
    if type(num) is int:
        return num
    cleaned_num = num.replace(',', '')
    cleaned_num = cleaned_num.replace('$', '')
    return int(cleaned_num)

# Clean numeric fields and convert to Integers
shows['Actual gross'] = shows['Actual gross'].apply(lambda x: clean_monetary_value(x) if pd.notna(x) and x != '' else x)
shows['Actual gross (in 2022 dollars)'] = shows['Actual gross (in 2022 dollars)'].apply(lambda x: clean_monetary_value(x) if pd.notna(x) and x != '' else x)
shows['Average gross'] = shows['Average gross'].apply(lambda x: clean_monetary_value(x) if pd.notna(x) and x != '' else x)
shows
```

### Imputation

Perform data imputation on the missing values in the dataset. Specifically:
- Row 6
- Row 11
- Row 19

...on the `Actual Gross`, `Actual Gross (in 2022 dollars)`, and `Average Gross` columns

```python
# Imputate values for rows at index 5, 10, and 18
# Actual gross values
shows.loc[5, 'Actual gross'] = 305_200_000
shows.loc[10, 'Actual gross'] = 229_700_000
shows.loc[18, 'Actual gross'] = 278_442_753

# Actual gross (in 2022 dollars) values
shows.loc[5, 'Actual gross (in 2022 dollars)'] = int(shows['Actual gross'][5] * 1.29)
shows.loc[10, 'Actual gross (in 2022 dollars)'] = int(shows['Actual gross'][10] * 1.20)
shows.loc[18, 'Actual gross (in 2022 dollars)'] = int(shows['Actual gross'][18] * 1.34)

# Average gross values
shows.loc[5, 'Average gross'] = int(shows['Actual gross (in 2022 dollars)'][5] / int(shows['Shows'][5]))
shows.loc[10, 'Average gross'] = int(shows['Actual gross (in 2022 dollars)'][10] / int(shows['Shows'][10]))
shows.loc[18, 'Average gross'] = int(shows['Actual gross (in 2022 dollars)'][18] / int(shows['Shows'][18]))
shows
```

### Further cleaning

It seems that there is an error in the values for the `Rank` column and `Adele Live 2016` should not be in 19th place after adding the actual values for the gross

```python
# Move Adele Tour to Rank 8 to fix ranking
shows = pd.concat([shows.iloc[:7], pd.DataFrame([shows.iloc[18]]), shows.iloc[7:]]).reset_index(drop=True)
shows = shows.drop(index=19).reset_index(drop=True)

# Fix Ranking values
for rank in range(19):
    shows.loc[rank, 'Rank'] = rank + 1

shows
```

### Normalization

We can now compute for the normalized values of the Gross columns (`Actual gross`, `Actual gross (in 2022 dollars)`, `Average gross`)

```python
# Compute Normalized Values
columns = ['Actual gross', 'Actual gross (in 2022 dollars)', 'Average gross']
normalized_shows = shows.copy()
normalized_shows[columns] = (normalized_shows[columns] - normalized_shows[columns].min()) / (normalized_shows[columns].max() - normalized_shows[columns].min())
normalized_shows
```

### Standardization
We can now compute for the standardized values of the Gross columns (`Actual gross`, `Actual gross (in 2022 dollars)`, `Average gross`)

```python
# Compute Standardized Values
standardized_shows = shows.copy()
standardized_shows[columns] = (standardized_shows[columns] - standardized_shows[columns].mean()) / standardized_shows[columns].std(ddof=0)
standardized_shows
```

### Discretization
We can now perform Discretization on the dataset by categorizing the Average Gross values

```python
discretized_shows = shows.copy()
discretized_shows = discretized_shows.drop(columns=['Actual gross', 'Actual gross (in 2022 dollars)'])

def categorize_avg_gross(num):
    if(num > 10_000_000):
        return 'Very Consistent Viewership'
    elif(num > 5_000_000):
        return 'Steady Viewership'
    elif(num > 2_000_000):
        return 'Variable Viewership'

    return 'Inconsistent Viewership'

discretized_shows = discretized_shows.assign(Category=discretized_shows['Average gross'].apply(categorize_avg_gross))
discretized_shows
```
